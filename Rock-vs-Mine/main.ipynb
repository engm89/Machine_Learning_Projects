#### Build a `Logistic Regression` Model to predict whether an object is Rock or Mine using SONAR Dataset.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset
data = pd.read_csv('./sonar_data.csv', header=None) # header=None --> there is no head/title in dataset
data
```

```python
data.info()
```

```python
data.head()
```

```python
data.shape
```

```python
data[60].value_counts()
```

```python
# sparating into features and label
X = data.drop(columns=60, axis=1)
y = data[60]
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)
```

```python
X.shape, X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
logistModel = LogisticRegression()
```

```python
logistModel
```

```python
logistModel.fit(X_train, y_train)
```

```python
# accuracy_score on Training Sets
training_prediction = logistModel.predict(X_train)
training_accuracy = accuracy_score(training_prediction, y_train)

test_predication = logistModel.predict(X_test)
test_accuracy = accuracy_score(test_predication, y_test)
```

```python
print(f'Training Accuracy:  {training_accuracy}, and Test Accuaracy: {test_accuracy}')
```
